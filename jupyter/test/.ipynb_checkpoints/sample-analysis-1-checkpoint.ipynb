```python
import pymongo
from pymongo import MongoClient
import pandas as pd
from pandas.io.json import json_normalize
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
client = MongoClient('localhost', 27017)
db = client.bitfeed
record = db["store_v1"].find_one()
```

```python
plt.bar([p for p,v in record["asks"]], height=[v for p,v in record["asks"]], width=0.5, color='red')
plt.bar([p for p,v in record["bids"]], height=[v for p,v in record["bids"]], width=0.5, color='green')
```

```python
cursor = db["store_v1"].find({"exchSym":"XBTUSD"}).limit(10000)
datapoints = []
for d in cursor:
    data = {"exchSym":d["exchSym"],
            "exchange":d["exchange"],
            "timestamp":d["timestamp"]}
    for i,(p,v) in enumerate(d['asks']):
        data['ask.p.%2d'%(i+1)] = p
        data['ask.v.%2d'%(i+1)] = v
    for i,(p,v) in enumerate(d['bids']):
        data['bid.p.%2d'%(i+1)] = p
        data['bid.v.%2d'%(i+1)] = v
    datapoints.append(data)
```

```python
df = pd.DataFrame(datapoints)
df.head()
```
